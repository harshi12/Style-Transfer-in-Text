```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./csv/Twitter-sentiment-self-drive-DFE.csv',encoding = 'unicode_escape')
```

```python
df
```

```python
df1 = df[['text','sentiment']]
df1 = df1[df1['sentiment'] != 'not_relevant']
# df1
```

```python
list_senti = pd.to_numeric(df1['sentiment'])
df1['sentiment'] = list_senti
df1 = df1[(df1['sentiment'] > 3) | (df1['sentiment'] < 3)]
```

```python
list_senti = []
for index, row in df1.iterrows():
    if row['sentiment'] > 3:
        list_senti.append(1)
    else:
        list_senti.append(0)

df1['sentiment'] = list_senti
 
```

```python
df1.to_csv('./csv/Twitter.txt', index = False , header = False, sep = ' ')
```
