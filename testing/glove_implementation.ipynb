## Inputting data

```python
neg_sent = pd.read_fwf('./sentiment/neg.txt')
neg_sent=neg_sent[0:500]
# pos_sent = pd.read_fwf('./sentiment/pos.txt')
```

```python
# neg_sent=np.reshape(neg_sent,)
neg_sent=neg_sent.values
neg_sent=neg_sent.flatten()

len(neg_sent)
for i in neg_sent:
    print (i)
    
```

## removing stopwords

```python
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 


stop_words = set(stopwords.words('english'))
neg_tokens_without_stopwords=[] #stop words contain the set of stop words 


for line in neg_sent: 
    temp_line=[] 
    word_tokens = word_tokenize(line)
    filtered_sentence = [w for w in word_tokens if not w in stop_words] 
    neg_tokens_without_stopwords.append(filtered_sentence)

neg_tokens=neg_tokens_without_stopwords
print(neg_tokens_without_stopwords)
```

```python
# import WordNet Lemmatizer from nltk 
from nltk.stem import WordNetLemmatizer 
wordnet_lemmatizer = WordNetLemmatizer() 
lemmatize_tokens=[] #stop words contain the set of stop words 

for line in neg_tokens: 
    temp_line=[] 
    for word in line:
        lemmatize_tokens.append (wordnet_lemmatizer.lemmatize(word)) 
# print(lemmatize_tokens)
neg_tokens=lemmatize_tokens
neg_tokens
```

## Using Glove on tokenized data

```python
#importing the glove library
from glove import Corpus, Glove
# creating a corpus object
corpus = Corpus()
print(type(neg_tokens))
#training the corpus to generate the co occurence matrix which is used in GloVe
corpus.fit(neg_tokens, window=20)
#creating a Glove object which will use the matrix created in the above lines to create embeddings
#We can set the learning rate as it uses Gradient Descent and number of components
glove = Glove(no_components=10, learning_rate=0.05)
 
glove.fit(corpus.matrix, epochs=30, no_threads=4, verbose=True)
glove.add_dictionary(corpus.dictionary)
glove.save('glove.model')

print(corpus.dictionary)
```
